#### Create New Feature Total Purchase 

```python
df['Total Purchases'] = df['Purchase Amount (USD)'] * df['Previous Purchases']
df
```

#### Top 10 Customers By Total Purchase Amount

```python

top_10_customers = df.sort_values(by='Total Purchases', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Customer ID', y='Total Purchases', data=top_10_customers, palette='viridis', order=top_10_customers.sort_values('Total Purchases', ascending=False)['Customer ID'])
plt.title('Top 10 Customers by Total Purchase Amount (Descending Order)')
plt.xlabel('Customer ID')
plt.ylabel('Total Purchases')
plt.show()
```

####  Counts Of Customers By Total  Purchase Amount 

```python
count_purchase = df['Customer ID'].count()
count_mid_purchase = df[(df['Total Purchases'] >= 2499) & (df['Total Purchases'] <= 4499)]['Customer ID'].count()
count_high_purchase = df[df['Total Purchases'] >= 4499]['Customer ID'].count()

print(f"Number of customers  {count_purchase}")
print(f"Number of customers with total purchase amount between  2499-4499: {count_mid_purchase}")
print(f"Number of customers with total purchase amount minimum 4499: {count_high_purchase}")
```

```python
#Label Encoding 
from sklearn.preprocessing import LabelEncoder

categorical_columns = ['Gender', 'Category', 'Location', 'Size', 'Color', 'Season', 'Subscription Status', 'Payment Method', 'Shipping Type', 'Discount Applied', 'Promo Code Used', 'Preferred Payment Method', 'Frequency of Purchases']

label_encoder = LabelEncoder()

for column in categorical_columns:
    df[column + '_encoded'] = label_encoder.fit_transform(df[column])

print(df.head())
df
```

#### Corrlation Matrix

```python

correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm",fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

#### 1.Visualizing Age Distribution: Count and Plot

```python
# Age distribution
age_bins = [0, 20, 30, 40, 50, 60,70,80]
age_labels = ['0-20', '21-30', '31-40', '41-50', '51-60','61-70','71-80']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
#This line of code creates a new column in the DataFrame called 'Age Group'. It assigns each row in the 'Age' column to a 
#specific age group based on the bins and labels defined earlier. The pd.cut() function is used for this purpose.
#right=False - left close and right open 
# Plotting age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=age_bins, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
```

#### 2.Count of Male and Female customers


```python
gender_counts = df['Gender'].value_counts() #counts the occurrences of each unique value in the 'Gender' 
print(gender_counts)
# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral']) #gender_counts.index retrieves the index (unique values) of the gender_counts
plt.title('Distribution of Male and Female Customers') #index contains the unique gender values
plt.show()
```

#### 3.Calculate total spend by gender

```python
gender_spend = df.groupby('Gender')['Purchase Amount (USD)'].sum()

plt.figure(figsize=(8, 6))
sns.barplot(x=gender_spend.index, y=gender_spend.values, palette='pastel')
plt.title('Total Spending by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Purchase Amount (USD)')
plt.show()
```

#### 4.Total Spending by Age Group

```python
# Define age groups
age_bins = [0, 20, 30, 40, 50, 60, 100]
age_labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '60+']

df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

age_group_spend = df.groupby('Age Group')['Purchase Amount (USD)'].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=age_group_spend.index, y=age_group_spend.values, palette='muted')
plt.title('Total Spending by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Purchase Amount (USD)')
plt.show()
```

#### 18. Find Subscription Status By Different Age Group

```python
df_yes = df[df['Subscription Status'] == 'Yes']

age_bins = [0, 20, 30, 40, 50, 60, 100] 
age_labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61+']  

df_yes['Age Group'] = pd.cut(df_yes['Age'], bins=age_bins, labels=age_labels, right=False)

subscription_counts = df_yes.groupby('Age Group').size().reset_index(name='Subscription Yes Count')

plt.figure(figsize=(10, 6))
sns.barplot(x='Age Group', y='Subscription Yes Count', data=subscription_counts, palette='viridis')
plt.title('Subscription Status - Yes Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count of Subscription Status - Yes')
plt.show()
```
